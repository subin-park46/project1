```python
#여러변수출력코드
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"

#모듈 import
import numpy as np
import pandas as pd
```

```python
gas2016=pd.read_csv('C:/ds_work/data/gas2016.csv')
gas2017=pd.read_csv('C:/ds_work/data/gas2017.csv')
gas2018=pd.read_csv('C:/ds_work/data/gas2018.csv')
gas2019=pd.read_csv('C:/ds_work/data/gas2019.csv')
gas2020=pd.read_csv('C:/ds_work/data/gas2019.csv')
```

```python
#위아래연결
res=pd.concat([gas2016,gas2017,gas2018,gas2019,gas2020],ignore_index=False)
res
```

```python
res.info()
```

```python
res.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
#연도별 최대 온실가스 배출
year_max=res.groupby('TRGT_YEAR')['GHG_EMS'].max()
year_max=year_max.reset_index()

plt.plot(year_max['TRGT_YEAR'], year_max['GHG_EMS'], color='orange', marker = 'o')
plt.show()
```
